```python
from scipy.stats.stats import pearsonr
```

```python
print("(No motion test)Type : (Pearson’s correlation coefficient, 2-tailed p-value)")
print("Accelerometer X :", pearsonr(still_andro['ACCELEROMETER X (m/s²)'], still_acc['x']))
print("Accelerometer Y :", pearsonr(still_andro['ACCELEROMETER Y (m/s²)'], still_acc['y']))
print("Accelerometer Z :", pearsonr(still_andro['ACCELEROMETER Z (m/s²)'], still_acc['z']))
print("Magnetometer X :", pearsonr(still_andro['MAGNETIC FIELD X (μT)'], still_mag['x']))
print("Magnetometer Y :", pearsonr(still_andro['MAGNETIC FIELD Y (μT)'], still_mag['y']))
print("Magnetometer Z :", pearsonr(still_andro['MAGNETIC FIELD Z (μT)'], still_mag['z']))
print("Gyroscope X :", pearsonr(still_andro['GYROSCOPE X (rad/s)'], still_gyro['x']))
print("Gyroscope Y :", pearsonr(still_andro['GYROSCOPE Y (rad/s)'], still_gyro['y']))
print("Gyroscope Z :", pearsonr(still_andro['GYROSCOPE Z (rad/s)'], still_gyro['z']))
print("GPS Lat :", pearsonr(still_andro['LOCATION Latitude : '], still_gps['lat']))
print("GPS Long :", pearsonr(still_andro['LOCATION Longitude : '], still_gps['long']))
```

```python
print("(Walking motion test)Type : (Pearson’s correlation coefficient, 2-tailed p-value)")
print("Accelerometer X :", pearsonr(walk_andro['ACCELEROMETER X (m/s²)'], walk_acc['x']))
print("Accelerometer Y :", pearsonr(walk_andro['ACCELEROMETER Y (m/s²)'], walk_acc['y']))
print("Accelerometer Z :", pearsonr(walk_andro['ACCELEROMETER Z (m/s²)'], walk_acc['z']))
print("Magnetometer X :", pearsonr(walk_andro['MAGNETIC FIELD X (μT)'], walk_mag['x']))
print("Magnetometer Y :", pearsonr(walk_andro['MAGNETIC FIELD Y (μT)'], walk_mag['y']))
print("Magnetometer Z :", pearsonr(walk_andro['MAGNETIC FIELD Z (μT)'], walk_mag['z']))
print("Gyroscope X :", pearsonr(walk_andro['GYROSCOPE X (rad/s)'], walk_gyro['x']))
print("Gyroscope Y :", pearsonr(walk_andro['GYROSCOPE Y (rad/s)'], walk_gyro['y']))
print("Gyroscope Z :", pearsonr(walk_andro['GYROSCOPE Z (rad/s)'], walk_gyro['z']))
print("GPS Lat :", pearsonr(walk_andro['LOCATION Latitude : '], walk_gps['lat']))
print("GPS Long :", pearsonr(walk_andro['LOCATION Longitude : '], walk_gps['long']))
```
